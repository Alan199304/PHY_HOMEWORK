```python
def location_match(location1): 
    
    all_location = {'United States':'US', 
                      'United Kingdom':'UK', 
                      'Switzerland':'CH', 
                      'Singapore':'SG',
       'China (Mainland)':'CN', 
                      'Hong Kong SAR':'HK', 
                      'Japan':'JP', 
                      'France':'FR'}
    
    try:
        return all_location[location1]
    except:
        return 'Not in the location list'
```
